# Wine data / Decision Tree

````python
import pandas as pd

red_url = 'https://raw.githubusercontent.com/PinkWink/ML_tutorial/master/dataset/winequality-red.csv'
white_url = 'https://raw.githubusercontent.com/PinkWink/ML_tutorial/master/dataset/winequality-white.csv'

red_wine = pd.read_csv(red_url, sep=';')
white_wine = pd.read_csv(white_url, sep=';')
````

````python
# 데이터 구족 같아서 구분할 수 있도록 구분해주는 컬럼 'color' 생성
red_wine['color'] = 1.
white_wine['color'] = 0.

wine = pd.concat([red_wine, white_wine])
wine.info()
````

## 레드 와인 / 화이트 와인 분류기
- Feature : 'color'제외 - 'X'로 할당
- Target : 'color' - 'y'로 할당

```
1. 머신러닝을 통하여 '분류를 위한 모델'의 성능을 확인하기 위해 train data, test data로 나누고
2. train data로 학습 후, test data로 accuracy score(performance evaluation)을 확인.
3. DecisionTreeClasifier로 만든 인스턴스 'wine_tree'로 'y_pred_tr(예측값)'을 만들어 acc score를 확인
```



````python
# X는 다차원 행렬 대문자, y는 1차원 행렬 소문자.
X = wine.drop(['color'], axis=1) 
y = wine['color']
````

````python
from sklearn.model_selection import train_test_split
import numpy as np

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.2, 
                                                    random_state=13)

np.unique(y_train, return_counts=True)
````

````python
'''

X_train - 5197 
X_test - 1300
y_train - 5197
y_test - 1300

'''

y_test.shape 
````

````python
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Histogram(x=X_train['quality'], name='Train'))
fig.add_trace(go.Histogram(x=X_test['quality'], name='Test'))

fig.update_layout(barmode='overlay') # 겹치도록
fig.update_traces(opacity=0.75)
fig.show()
````

````python
from sklearn.tree import DecisionTreeClassifier

wine_tree = DecisionTreeClassifier(max_depth=2, random_state=13)
wine_tree.fit(X_train, y_train)
````

````python
from sklearn.metrics import accuracy_score

y_pred_tr = wine_tree.predict(X_train)
y_pred_test = wine_tree.predict(X_test)

print('Train Acc : ', accuracy_score(y_train, y_pred_tr))
print('Test Acc : ', accuracy_score(y_test, y_pred_test))

# Train data와 Test data는 유사하다.
````

## 데이터 전처리 - MinMaxScaler 와 StandardScaler

````python
fig = go.Figure()
fig.add_trace(go.Box(y=X['fixed acidity'], name='fixed acidity'))
fig.add_trace(go.Box(y=X['chlorides'], name='chlorides'))
fig.add_trace(go.Box(y=X['quality'], name='quality'))

fig.show()
````

````python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

MMS = MinMaxScaler()
SS = StandardScaler()

SS.fit(X)
MMS.fit(X)

X_ss = SS.transform(X)
X_mms = MMS.transform(X)
````

````python
X_ss_pd = pd.DataFrame(X_ss, columns=X.columns)
X_mms_pd = pd.DataFrame(X_mms, columns=X.columns)
````

````python
fig = go.Figure()
fig.add_trace(go.Box(y=X_mms_pd['fixed acidity'], name='fixed acidity'))
fig.add_trace(go.Box(y=X_mms_pd['chlorides'], name='chlorides'))
fig.add_trace(go.Box(y=X_mms_pd['quality'], name='quality'))

fig.show()
````

````python
fig = go.Figure()
fig.add_trace(go.Box(y=X_ss_pd['fixed acidity'], name='fixed acidity'))
fig.add_trace(go.Box(y=X_ss_pd['chlorides'], name='chlorides'))
fig.add_trace(go.Box(y=X_ss_pd['quality'], name='quality'))

fig.show()
````

````python
X_train, X_test, y_train, y_test = train_test_split(X_mms_pd, y,
                                                    test_size=0.2,
                                                    random_state=13
                                                    )

wine_tree = DecisionTreeClassifier(max_depth=2, random_state=13)
wine_tree.fit(X_train, y_train)

y_pred_tr = wine_tree.predict(X_train)
y_pred_test = wine_tree.predict(X_test)

print('Train Acc : ', accuracy_score(y_train, y_pred_tr))
print('Test Acc : ', accuracy_score(y_test, y_pred_test))
````

````python
X_train, X_test, y_train, y_test = train_test_split(X_ss_pd, y,
                                                    test_size=0.2,
                                                    random_state=13
                                                    )

wine_tree = DecisionTreeClassifier(max_depth=2, random_state=13)
wine_tree.fit(X_train, y_train)

y_pred_tr = wine_tree.predict(X_train)
y_pred_test = wine_tree.predict(X_test)

print('Train Acc : ', accuracy_score(y_train, y_pred_tr))
print('Test Acc : ', accuracy_score(y_test, y_pred_test))
````
