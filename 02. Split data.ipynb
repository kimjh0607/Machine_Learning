```python
from sklearn.datasets import load_iris
import pandas as pd
iris = load_iris()
```

```python
from sklearn.model_selection import train_test_split

features = iris.data[:, 2:]
labels = iris.target
```

```python
# 테스트 데이터는 20%로 설정하겠다.
X_train, X_test, y_train, y_test = train_test_split(features, labels, 
                                                    test_size=0.2, 
                                                    stratify=labels,
                                                    random_state=13) 
```

```python
X_train, X_train.shape
```

```python
X_test, X_test.shape
```

```python
y_test, y_test.shape
```

```python
y_train, y_train.shape
```

```python
import numpy as np

np.unique(y_test, return_counts=True)
```

```python
from sklearn.tree import DecisionTreeClassifier

iris_tree = DecisionTreeClassifier(max_depth=2, random_state=13)
iris_tree.fit(X_train, y_train)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
plot_tree(iris_tree)
```

```python
# train 데이터에 대한 accuracy 를 확인.
from sklearn.metrics import accuracy_score

y_pred_tr = iris_tree.predict(X_train)

accuracy_score(y_train, y_pred_tr)
```

```python
# 훈련용 데이터에 대한 결정경계
from mlxtend.plotting import plot_decision_regions

plt.figure(figsize=(14,8))
plot_decision_regions(X=X_train, y=y_train, clf=iris_tree, legend=2)
plt.show()
```

```python
y_pred_test = iris_tree.predict(X_test)
accuracy_score(y_test, y_pred_test)
```

```python
plt.figure(figsize=(14,8))
plot_decision_regions(X=X_test, y=y_test, clf=iris_tree, legend=2)
plt.show()
```

```python
scatter_highlight_kwargs = {'s':150, 'label':'Test data', 'alpha':0.9}
# size 150, label지정, 투명도 설정
scatter_kwargs = {'s':120, 'edgecolor':None, 'alpha':0.9}

plt.figure(figsize=(12,8))
plot_decision_regions(X=features, y=labels,
                      X_highlight=X_test,
                      clf=iris_tree,
                      legend=2,
                      scatter_highlight_kwargs=scatter_highlight_kwargs,
                      scatter_kwargs=scatter_kwargs,
                      contourf_kwargs={'alpha':0.2}
                      )
```

```python
features = iris.data
labels = iris.target

X_train, X_test, y_train, y_test = train_test_split(features, labels, 
                                                    test_size=0.2, 
                                                    stratify=labels,
                                                    random_state=13) 

iris_tree = DecisionTreeClassifier(max_depth=2, random_state=13)
iris_tree.fit(X_train, y_train)
```

```python
plt.figure(figsize=(12,8))
plot_tree(iris_tree)
```

## zip과 unpacking

```python
# zip을 이용, list를 tuple로
list1 = ['a', 'b', 'c']
list2 = [1, 2, 3]

pairs = [pair for pair in zip(list1, list2)]
pairs
```

```python
# tuple을 dict로
dict(pairs)
```

```python
dict(zip(list1, list2))
```

```python
# unpacking 인자를 이용한 역변환
x, y = zip(*pairs)
```

```python
# 튜플로 나오게 된다.
x
```
